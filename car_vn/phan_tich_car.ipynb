```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_csv('/Users/binh230901/Desktop/DA/project/miai/car_vn/data/car.csv', delimiter='|',engine='python',error_bad_lines=False,warn_bad_lines=False, index_col=False)
```

## so sánh loại xe nào chiếm nhiều nhất

```python
# đếm số lượng xe của từng loại
car_type_count = df.groupby(df['car_type'])['car_model'].count().reset_index()
```

```python
# top 5 loại xe có số lượng nhiều nhất
car_type_count.sort_values(by='car_model', inplace= True, ascending= False )
car_type_count = car_type_count[0:5]
car_type_count
```

```python
# vẽ biểu đồ
plt.style.use('seaborn-whitegrid')
plt.figure(figsize= (10,10), dpi= 80)
plt.pie(car_type_count['car_model'],
        labels= car_type_count['car_type'],
        autopct='%.2f%%',
        explode= [0.1, 0.1, 0, 0, 0]
       )

plt.title('tỷ lệ các loại xe',
          fontsize= 15,
          loc= 'center'
         )

plt.legend(loc= 'upper right',
           ncol= 1
          );
```

## top hãng xe được ưa chuộng nhất

```python
# xử lý dữ liệu
df['automaker'] = df.loc[:,'car_model']
# tách hãng xe với model được cách nhau bởi step
df['automaker'] = df['automaker'].apply(lambda x: x.split(sep= '-')[0])
df['automaker']
```

```python
# kiểm tra xem đâu là hãng xe chiếm nhiều nhất
automaker = df.groupby(df['automaker'])['car_model'].count().reset_index(name= 'count')
automaker.sort_values(by='count', ascending= False, inplace= True)
automaker = automaker[0:10]
automaker
```

```python
# vẽ biểu đồ
plt.figure(figsize= (20,10))
plt.bar(x= automaker['automaker'], 
        height= automaker['count']
       )
# hiển thị số lượng trên mỗi bar của biểu đồ
for x, y in enumerate(automaker['count']):
    plt.text(x, y, y,
             ha= 'center', 
             va= 'bottom',
             fontsize= 12
            )
# tùy chỉnh các thông số khác
plt.xlabel('Hãng xe', fontsize= 15)
plt.xticks(fontsize= 12)
plt.ylabel('số lượng', fontsize= 15)
plt.yticks(fontsize= 12)
plt.title('top hãng xe được ưa chuộng', fontsize= 20)
```

không quá bất ngờ khi mẫu xe chiếm nhiều nhất là toyota và điều khá thú vị là trong top5 này có mẫu xe đến từ nhà mercedes một hãng xe hạng sang mà vẫn đứng được trong top5 phổ biến nhất, nể phục 

## màu xe được ưa thích

## biểu đồ histogram theo mức giá

```python
# xử lý tiền dữ liệu
# chuyển giá ví dụ: 1 tỷ 400 thành 1400000 để có thể tính toán và sắp xếp đc 
# tách ra thành từng phần tử ngăn cách nhau bởi dấu cách
price = df['price'].apply(lambda x: x.split())
# chạy vòng lặp duyệt qua hết các phần tử 
for x in range(len(price)):
    if len(price[x]) == 2:
        price[x] = int(price[x][0])*1000
    elif len(price[x]) == 4:
        price[x] = int(price[x][0])*1000000 + int(price[x][2]) *1000
df['price_convert']= price
# convert to numeric
df['price_convert']= pd.to_numeric(df['price_convert'])
df
```

```python
# vẽ biểu đồ histogram
bins = list(range(0, 2000000, 50000))
plt.figure(figsize= (20, 10))
plt.style.use('seaborn-whitegrid')
plt.hist(x= df['price_convert'], bins= bins , edgecolor= 'black', rwidth= 0.85)
plt.title('số lượng xe theo từng mức giá', fontsize= 20)
plt.xlabel('mức giá (1 tỷ đồng)', fontsize= 15)
plt.ylabel('số lượng xe', fontsize= 15)
# tính percentiles 
pctiles = df['price_convert'].quantile(0.8)
plt.axvline(x= pctiles, color= 'red', label= '80% percentiles', linestyle = '--')
plt.legend(fontsize = 20);
```

hơn 80% giao dịch là các xe có giá dưới 1 tỷ

từ 250 - 600 triệu có số lượng tin đăng bán nhiều nhất (phổ biến nhất) 

# người ta thường nghĩ tới việc bán xe sau khi đi được bao nhiêu km nhỉ ??

```python
# convert km ví dụ 5 km thành 5
for i in range(len(df['km'])):
    df['km_1'][i] = df['km'][i].split()[0].replace(',', '') 
df['km_1'] = pd.to_numeric(df['km_1'])
df.head()
```

```python
# lọc ra xe cũ
old_car = df.loc[(df['km_1'] > 0), ['automaker','km_1','new_old' ]]
old_car
```

```python
# vẽ biểu đồ hist
plt.figure(figsize= (20, 10))
plt.hist(x= old_car['km_1'],bins = list(range(0, 200000, 10000)),rwidth= 0.9 , edgecolor= 'black',histtype='barstacked' )

plt.xlabel('số km đi được', fontsize= 15)
plt.ylabel('số lượng xe', fontsize= 15);
# vẽ vline :
vline = old_car['km_1'].quantile(0.91)
plt.axvline(x= vline, color= 'red', label= '90% percentile', linestyle= '--')
plt.legend(fontsize= 20);
```

sự xuất hiện nhiều thanh bar cao bất thường như vậy là mình nghĩ do người dùng thường làm tròn số km nên ở các số km như 50000, 60000, 70000,... thường tập trung nhiều xe bán hơn và nhìn chung thì  tới 90% lượng xe bán ra là từ 150000km trở xuống
